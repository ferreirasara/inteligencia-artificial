```python
from copy import deepcopy
```

```python
def isValidMove(state, origin, destiny):
    # the source tower disk must be smaller than the destination tower disk
    if state[origin] == []:
        return False
    elif state[destiny] == []:
        return True
    else:
        return state[origin][-1] < state[destiny][-1]
```

```python
def isObjectiveState(state):
    tower1 = state[0] == []
    tower2 = state[1] == []
    tower3 = state[2] == [5,4,3,2,1]
    
    return tower1 and tower2 and tower3
```

```python
def action(state, origin, destiny):
    newState = deepcopy(state)
    disk = newState[origin].pop()
    newState[destiny].append(disk)
    
    return newState
```

```python
def expand(state):
    moves = []
    for i in range(3):
        for j in range(3):
            if i != j:
                if isValidMove(state, i, j):
                    moves.append(action(state, i, j))
                    
    return moves
```

```python
def apeendStatesToFringe(fringe, states):
    for state in states:
        fringe.append(state)
```

```python
def printSolution(solution):
    for state in solution:
        print(state[0], state[1], state[2])
```

```python
def treeSearch(fringe, visitedStates, solution):
    if fringe == []:
        print("Fringe empty!")
        return False
    else:
        currentState = fringe.pop(0)
        if currentState not in visitedStates:
            visitedStates.append(currentState)
            if isObjectiveState(currentState):
                solution.append(currentState)
                print(f"FINAL! Resolved in {len(solution)} movements.")
                printSolution(solution)
                return True
            else:
                apeendStatesToFringe(fringe, expand(currentState))
                for state in fringe:
                    try:
                        solution.remove(state)
                    except:
                        pass
                solution.append(currentState)
                treeSearch(fringe, visitedStates, solution)
        else:
            solution.append(currentState)
            treeSearch(fringe, visitedStates, solution)
```

```python
initialState = [
    [5,4,3,2,1],
    [],
    []
]

fringe = [initialState]
visitedStates = []
solution = []

treeSearch(fringe, visitedStates, solution)
```
