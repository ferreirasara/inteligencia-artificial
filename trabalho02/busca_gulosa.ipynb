```python
def checkPath(maze, x, y):
    return (maze[x][y] == ' ' or maze[x][y] == 'S')
```

```python
def possiblePaths(maze, x, y):
    possiblePaths = []
    if checkPath(maze, x-1, y):
        possiblePaths.append([x-1, y])
    if checkPath(maze, x+1, y):
        possiblePaths.append([x+1, y])
    if checkPath(maze, x, y-1):
        possiblePaths.append([x, y-1])
    if checkPath(maze, x, y+1):
        possiblePaths.append([x, y+1])
    return possiblePaths
```

```python
def markPath(maze, x, y):
    maze[x][y] = '0'
```

```python
def chooseNextPath(maze, paths, destiny):
    nextPath = {
        'path': [0,0],
        'cost': 999
    }
    for path in paths:
#         cost = [destiny[0] - path[0], destiny[1] - path[1]]
        cost = (destiny[0] - path[0]) + (destiny[1] - path[1])
#         if cost[0] <= nextPath['cost'][0] or cost[1] <= nextPath['cost'][1]:
        if cost <= nextPath['cost']:
            nextPath['path'] = path
            nextPath['cost'] = cost
            
    return nextPath['path']
```

```python
def greedySearch(maze, origin, destiny):
    x, y = origin
    while True:
        print(x,y)
        if isExit(maze, x, y):
            print("Exit!")
            return
        else:
            paths = possiblePaths(maze, x, y)
            if paths == []:
                print("Dead End!")
                return
            elif len(paths) == 1:
                x = paths[0][0]
                y = paths[0][1]
                markPath(maze, x, y)
            else:
                nextPath = chooseNextPath(maze, paths, destiny)
                x = nextPath[0]
                y = nextPath[1]
                markPath(maze, x, y)
```
