```python
from math import fabs, pow, sqrt
```

```python
def fileToMatrix(fileName):
    file = open(fileName, 'r')
    matrix = []
    
    for fileLine in file:
        mazeLine = []
        mazeLine[:0] = fileLine
        
        matrix.append(mazeLine[:-1])
    
    file.close()
    
    return matrix
```

```python
def printMaze(maze):
    for line in maze:
        for letter in line:
            print(letter, end = " ")
        print()
```

```python
def findEntrance(maze):
    for x in range(len(maze)):
        for y in range(len(maze[0])):
            if maze[x][y] == 'E':
                return x,y
```

```python
def findExit(maze):
    for x in range(len(maze)):
        for y in range(len(maze[0])):
            if maze[x][y] == 'S':
                return x,y
```

```python
def checkPath(maze, x, y):
    if x < 0 or x >= len(maze):
        return False
    elif y < 0 or y >= len(maze[0]):
        return False
    else:
        return (maze[x][y] == ' ') or (maze[x][y] == 'S')
```

```python
def possiblePaths(maze, x, y):
    possiblePaths = []
    if checkPath(maze, x-1, y):
        possiblePaths.append([x-1, y])
    if checkPath(maze, x+1, y):
        possiblePaths.append([x+1, y])
    if checkPath(maze, x, y-1):
        possiblePaths.append([x, y-1])
    if checkPath(maze, x, y+1):
        possiblePaths.append([x, y+1])
    return possiblePaths
```

```python
def markPath(maze, x, y):
    maze[x][y] = '0'
```

```python
def euclideanDistance(origin, destiny):
    x = origin[0] - destiny[0]
    y = origin[1] - destiny[1]
    z = pow(x, 2) + pow(y, 2)
    return sqrt(z)
```

```python
def chooseNextPath(maze, paths, destiny, actualCost):
    nextPath = {
        'path': 0,
        'cost': euclideanDistance(origin, destiny)
    }
    for path in paths:
        cost = euclideanDistance(path, destiny)
        if cost <= cost * actualCost and cost <= nextPath['cost']:
            nextPath['path'] = paths.index(path)
            nextPath['cost'] = cost
        return nextPath['path']
```

```python
def aStar(maze, origin, destiny):
    x, y = origin
    checkpoints = []
    checkpointsVisiteds = []
    actualCost = 0
    while True:
        if x == destiny[0] and y == destiny[1]:
            print("Exit!")
            return
        else:
            paths = possiblePaths(maze, x, y)
            
            if paths == []:
                if checkpoints == []:
                    print("Dead End!")
                    return
                checkpoint = checkpoints.pop(0)
                while checkpoint in checkpointsVisiteds:
                    checkpoint = checkpoints.pop(0)
                print(f"Dead End! Returning to checkpoint {checkpoint}")
                x = checkpoint[0]
                y = checkpoint[1]
                checkpointsVisiteds.append(checkpoint)
            elif len(paths) == 1:
                x = paths[0][0]
                y = paths[0][1]
                actualCost += 1
                markPath(maze, x, y)
            else:
                indexNextPath = chooseNextPath(maze, paths, destiny, actualCost)
                actualCost += 1
                nextPath = paths.pop(indexNextPath)
                x = nextPath[0]
                y = nextPath[1]
                checkpoints.extend(paths)
                markPath(maze, x, y)
```

```python
maze = fileToMatrix("maze04.txt")

origin = findEntrance(maze)
destiny = findExit(maze)

aStar(maze, origin, destiny)
printMaze(maze)
```
