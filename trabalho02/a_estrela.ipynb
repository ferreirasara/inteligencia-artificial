```python
from math import fabs, pow, sqrt
```

```python
def fileToMatrix(fileName):
    file = open(fileName, 'r')
    matrix = []
    
    for fileLine in file:
        mazeLine = []
        mazeLine[:0] = fileLine
        
        matrix.append(mazeLine[:-1])
    
    file.close()
    
    return matrix
```

```python
def printMaze(maze):
    for line in maze:
        for letter in line:
            print(letter, end = " ")
        print()
```

```python
def findEntrance(maze):
    for x in range(len(maze)):
        for y in range(len(maze[0])):
            if maze[x][y] == 'E':
                return x,y
```

```python
def findExit(maze):
    for x in range(len(maze)):
        for y in range(len(maze[0])):
            if maze[x][y] == 'S':
                return x,y
```

```python
def checkPath(maze, x, y):
    if x < 0 or x >= len(maze):
        return False
    elif y < 0 or y >= len(maze[0]):
        return False
    else:
        return (maze[x][y] == ' ') or (maze[x][y] == 'S')
```

```python
def possiblePaths(maze, x, y):
    possiblePaths = []
    if checkPath(maze, x-1, y):
        possiblePaths.append([x-1, y])
    if checkPath(maze, x+1, y):
        possiblePaths.append([x+1, y])
    if checkPath(maze, x, y-1):
        possiblePaths.append([x, y-1])
    if checkPath(maze, x, y+1):
        possiblePaths.append([x, y+1])
    return possiblePaths
```

```python
def markPath(maze, x, y):
    maze[x][y] = '0'
```

```python
def chooseNextPath(maze, paths, destiny, actualCost):
    nextPath = {
        'path': [0,0],
        'cost': 999999
    }
    for path in paths:
        x = destiny[0] - path[0]
        y = destiny[1] - path[1]
        z = pow(x, 2) + pow(y, 2)
        cost = sqrt(z)
        if cost + actualCost <= nextPath['cost']:
            nextPath['path'] = path
            nextPath['cost'] = cost + actualCost
    return nextPath['path'], nextPath['cost']
```

```python
def aStar(maze, origin, destiny):
    x, y = origin
    actualCost = 0
    while True:
        if x == destiny[0] and y == destiny[1]:
            print("Exit!")
            return
        else:
            paths = possiblePaths(maze, x, y)
            
            if paths == []:
                print("Dead End!")
                return
            elif len(paths) == 1:
                x = paths[0][0]
                y = paths[0][1]
                actualCost += 1
                markPath(maze, x, y)
            else:
                nextPath, actualCost = chooseNextPath(maze, paths, destiny, actualCost)
                x = nextPath[0]
                y = nextPath[1]
                markPath(maze, x, y)
```

```python
maze = fileToMatrix("maze03.txt")

origin = findEntrance(maze)
destiny = findExit(maze)

aStar(maze, origin, destiny)
printMaze(maze)
```
