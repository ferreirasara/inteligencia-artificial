# Hill Climbing with Random Restart

```python
def fileToMatrix(filename):
    file = open(filename, 'r')
    matrix = []
    
    for fileLine in file:
        line = fileLine.strip().split()
        if line[0].isnumeric():
            matrix.append(line)
    
    file.close()
    return matrix
```

```python
def getDistance(matrix, x, y):
    try:
        return matrix[x][y]
    except IndexError:
        return None
```
