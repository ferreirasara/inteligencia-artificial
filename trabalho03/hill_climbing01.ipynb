# Steepest-Ascent Hill climbing

```python
def fileToMatrix(filename):
    file = open(filename, 'r')
    matrix = []
    
    for fileLine in file:
        line = fileLine.strip().split()
        if line[0].isnumeric():
            matrix.append(line)
    
    file.close()
    return matrix
```

```python
matrix = fileToMatrix('br17.atsp')
print(matrix)
```
