```python
from random import randrange
from math import sqrt
```

```python
def calcRoots(a, b, c):
    delta = (b**2) - (4 * a * c)
#     x1 = (-b + delta**(1.0/2.0)) / (2.0*a)
#     x2 = (-b - delta**(1.0/2.0)) / (2.0*a)
    
    if delta < 0:
        return 'A equação não possui raizes reais'
    elif delta == 0:
        x1 = (-1 * b + sqrt(delta)) / (2 * a)
        return x1
    else:
        x1 = (-1 * b + sqrt(delta)) / (2 * a)
        x2 = (-1 * b - sqrt(delta)) / (2 * a)   
        return x1, x2

for i in range(5):
    a = randrange(-10, 10)
    b = randrange(-10, 10)
    c = randrange(-10, 10)

    print(f'{a}x^2 + {b}x + {c} = 0 : x = {calcRoots(a, b, c)}')
```

```python
def celsiusToFahrenheit(temperature):
    return ((temperature * 9/5) + 32)

for i in range(101):
    print(f'{i}°C = {celsiusToFahrenheit(i)}°F')
```

```python
def reverseString(string):
    # Using python string method
#     reversedString = [letter for letter in string]
#     reversedString.reverse()
#     return ''.join(reversedString)
    reversedString = []
    for letter in string:
        reversedString.insert(0,letter)
        
    return ''.join(reversedString)

print(reverseString("Sara"))
print(reverseString("Eisenbahn"))
print(reverseString("Atenas"))
print(reverseString("Amora"))
```

```python
def isPalindrome(string):
    reversedString = reverseString(string)
    return reversedString.lower() == string.lower()

print(isPalindrome("Ovo"))
print(isPalindrome("abba"))
print(isPalindrome("Beterraba"))
print(isPalindrome("SocorramMeSubiNoOnibusEmMarrocos"))
```

```python
class Abajur:
    def __init__(self):
        self.status = 1
    def on(self):
        self.status = 0
    def off(self):
        self.status = 1
    def __str__(self):
        return f'{"Ligado" if self.status == 0 else "Desligado"}'
    
abajur1 = Abajur()
abajur2 = Abajur()
abajur3 = Abajur()

print(f'Abajur 1: {abajur1} | Abajur 2: {abajur2} | Abajur 3: {abajur3}')

abajur1.on()
abajur2.on()
abajur3.on()

print(f'Abajur 1: {abajur1} | Abajur 2: {abajur2} | Abajur 3: {abajur3}')

abajur1.off()
abajur2.off()
abajur3.off()

print(f'Abajur 1: {abajur1} | Abajur 2: {abajur2} | Abajur 3: {abajur3}')
```

```python
class ContaCorrente:
    def __init__(self, name, number):
        self.name = name
        self.number = number
        self.budget = 0.00
    def depositar(self, value):
        self.budget += value
    def retirar(self, value):
        self.budget -= value
    def transferir(self, value, destinatioAccount):
        destinatioAccount.depositar(value)
        self.retirar(value)
    def extrato(self):
        print(f'Cliente: {self.name} | Saldo: {self.budget}')
    def __str__(self):
        return number
    
conta1 = ContaCorrente("Sara", "00001")
conta2 = ContaCorrente("Luciano", "00002")
conta3 = ContaCorrente("Luke", "00003")

print("Inicialização das contas".center(50, "-"))
conta1.extrato()
conta2.extrato()
conta3.extrato()

conta1.depositar(1060)
conta2.depositar(2030)
conta3.depositar(50321)

print("Depósitos".center(50, "-"))
conta1.extrato()
conta2.extrato()
conta3.extrato()

conta1.retirar(57)
conta2.retirar(1633)
conta3.retirar(823)

print("Retiradas".center(50, "-"))
conta1.extrato()
conta2.extrato()
conta3.extrato()

conta1.transferir(500, conta2)
conta2.transferir(300, conta1)
conta3.transferir(150, conta1)

print("Transferências".center(50, "-"))
conta1.extrato()
conta2.extrato()
conta3.extrato()
```

```python
class Retangulo:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def desenhar(self):
        for w in range(self.width):
            for h in range(self.height):
                print('*', end =" ")
            print()
        print()
    
retangulo1 = Retangulo(13, 15)
retangulo2 = Retangulo(10, 6)

retangulo1.desenhar()
retangulo2.desenhar()
```
